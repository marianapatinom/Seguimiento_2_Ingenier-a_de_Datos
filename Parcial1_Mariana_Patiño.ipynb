# Selección del Dataset

**Nombre del dataset:**  
Airline Disruptions Dataset

**Fuente:**  
Kaggle – Airline Disruptions and Operational Impact Dataset

**Descripción del dataset**

Este dataset contiene información sobre interrupciones operativas en aerolíneas, incluyendo vuelos cancelados, vuelos redirigidos, pasajeros afectados y pérdidas económicas estimadas.  
Los datos permiten analizar el impacto de eventos operacionales en diferentes aerolíneas y países.  
También se incluyen costos adicionales de combustible generados por redireccionamientos de vuelos.

Este dataset es útil para analizar patrones de interrupciones y su impacto económico en la industria aérea.

| Columna | Descripción |
|-------|------|
| airline | Aerolínea afectada |
| country | País donde ocurre la interrupción |
| estimated_daily_loss_usd | Pérdida económica diaria estimada |
| cancelled_flights | Número de vuelos cancelados |
| rerouted_flights | Número de vuelos redirigidos |
| additional_fuel_cost_usd | Costos adicionales de combustible |
| passengers_impacted | Número de pasajeros afectados |

# Proceso ETL

## Extracción

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("airline_losses_estimate.csv")
```

## Información general del dataset

El dataset contiene información sobre interrupciones operativas en aerolíneas, incluyendo cancelaciones de vuelos, redireccionamientos, pasajeros afectados y pérdidas económicas estimadas.
Estos datos permiten analizar el impacto de eventos operacionales en diferentes aerolíneas y países.

# Transformación

```python
df.isnull().sum()
```

Se revisaron los valores faltantes en cada columna para identificar posibles problemas en la calidad de los datos.
Este paso es importante para decidir si se deben eliminar o reemplazar valores faltantes.

## Tratamiento de datos faltantes

```python
df = df.dropna()
```

Los registros con valores faltantes fueron eliminados utilizando dropna() para asegurar que el análisis se realice con datos completos.
Esto ayuda a evitar errores o sesgos en el análisis posterior.

# Conversión de tipos de datos

```python
df["cancelled_flights"] = df["cancelled_flights"].astype(int)
df["passengers_impacted"] = df["passengers_impacted"].astype(int)
```

Se convirtieron algunas columnas a tipo entero para asegurar consistencia en los datos numéricos y permitir cálculos estadísticos adecuados.

 ## Creación de variable derivada

```python
df["impact_level"] = np.where(
    df["cancelled_flights"] > df["cancelled_flights"].median(),
    "High Impact",
    "Low Impact"
)
```

Se creó una nueva variable llamada **impact_level** que clasifica los eventos según el número de vuelos cancelados.
Esto permite distinguir entre interrupciones de alto impacto y bajo impacto.

# Cargar


```python
df.to_csv("cleaned_dataset.csv", index=False)
```

El dataset limpio se guardó como **cleaned_dataset.csv**.

Este dataset es más adecuado para análisis porque:
- No contiene valores faltantes.
- Los tipos de datos están correctamente definidos.
- Incluye una nueva variable que facilita el análisis del impacto de las interrupciones.

# Histograma

```python
fig = px.histogram(
    df,
    x="cancelled_flights",
    color="impact_level",
    nbins=20,
    title="Distribución de vuelos cancelados",
    marginal="box"
)

fig.show()
```

El histograma muestra cómo se distribuyen los vuelos cancelados en el dataset.
También permite observar la diferencia entre eventos de alto impacto y bajo impacto.

# Diagrama de dispersión

```python
fig = px.scatter_geo(
    df,
    locations="country",
    locationmode="country names",
    color="estimated_daily_loss_usd",
    size="passengers_impacted",
    hover_name="airline",
    hover_data=["cancelled_flights","rerouted_flights"],
    projection="natural earth",
    color_continuous_scale="Turbo",
    template="plotly_dark",
    title="Mapa global del impacto de interrupciones en aerolíneas"
)

fig.show()
```



Este mapa interactivo muestra el impacto de interrupciones en aerolíneas a nivel mundial.

El color representa las pérdidas económicas estimadas, mientras que el tamaño de los puntos indica la cantidad de pasajeros afectados.  
Esto permite identificar qué regiones presentan eventos de mayor impacto operativo y económico.

# Boxplot

```python
fig = px.box(
    df,
    x="country",
    y="cancelled_flights",
    color="country",
    title="Distribución de vuelos cancelados por país"
)

fig.show()
```

El boxplot permite comparar la distribución de vuelos cancelados entre diferentes países.
También permite identificar posibles valores atípicos.

# Gráfico adicional (barras)

```python
airline_counts = df["airline"].value_counts().reset_index()
airline_counts.columns = ["airline","count"]

fig = px.bar(
    airline_counts,
    x="airline",
    y="count",
    color="airline",
    title="Eventos de interrupción por aerolínea"
)

fig.show()
```

Este gráfico permite identificar qué aerolíneas presentan mayor número de interrupciones operativas.
Esto puede indicar diferencias en la frecuencia de eventos entre compañías.
